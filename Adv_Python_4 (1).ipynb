```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

img = Image.open(r"C:\Users\surya\images.jpg")
im = np.asarray(img)

# Original
plt.figure(figsize=(7,5))
plt.imshow(im)
plt.axis('off')
plt.show()

# RGB Channels
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(14,5))

ax1.imshow(im[:, :, 0], cmap='Reds_r'); ax1.set_title("Red"); ax1.axis('off')
ax2.imshow(im[:, :, 1], cmap='Greens_r'); ax2.set_title("Green"); ax2.axis('off')
ax3.imshow(im[:, :, 2], cmap='Blues_r'); ax3.set_title("Blue"); ax3.axis('off')

plt.show()

# Histogram
plt.figure(figsize=(8,5))
plt.hist(im.ravel(), bins=256, color='gray')
plt.show()
```

```python
import pandas as pd

arrays = [
    ['A', 'A', 'B', 'B'],
    ['Math', 'Science', 'Math', 'Science']
]

index = pd.MultiIndex.from_arrays(arrays, names=('Alphabet', 'Subject'))
s = pd.Series([90, 85, 88, 92], index=index)
print(s)
```

```python
import pandas as pd

tuples = [('A','Math'), ('A','Science'), ('B','Math'), ('B','Science')]
index = pd.MultiIndex.from_tuples(tuples, names=('Alphabet','Subject'))

s = pd.Series([90, 85, 88, 92], index=index)
print(s)
```

```python
import pandas as pd

index = pd.MultiIndex.from_product(
    [['A','B'], ['Math','Science']],
    names=('Alphabet','Subject')
)

s = pd.Series([90, 85, 88, 92], index=index)
print(s)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Alphabet': ['A','A','B','B'],
    'Subject': ['Math','Science','Math','Science']
})

index = pd.MultiIndex.from_frame(df)
s = pd.Series([90, 85, 88, 92], index=index)
print(s)
```
