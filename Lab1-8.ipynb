```python
odds = [1, 3, 5, 7]

# Assign each number in the list to a separate variable
a, b, c, d = odds
print(a)  # Shows: 1
print(b)
print(c)
print(d)
```

```python
odds = [1, 3, 5, 7]

# Store the first two numbers in 'a' and 'b'
# The remaining numbers go into a new list called 'c'
a, b, *c = odds
print(a)
print(b)
print(c)
```

```python
# First value → 'a', last value → 'c'
# Everything in the middle → list 'b'
nums = [3, 5, 7, 9]

a, *b, c = nums
print(a)  # 3
print(b)  # [5, 7]
print(c)  # 9
```

```python
# Function to add 3 numbers together
def a(x, y, z):
    return (x + y + z)

odds = [1, 3, 5]
print(a(*odds))
```

```python
# This function accepts many names and greets each one
def fun(*names):
    for n in names:
        print(f"hello,{n}")

fun("suriya", "hari", "prabu")
```

```python
# Dictionary with two pieces of information
person = {"name": "kumari", "age": 20}

# Pull out the values in order
name, age = person.values()

print(name)
print(age)
```

```python
# Create an empty dictionary and add values one by one
dic = {}
dic["name"] = "suriya"
dic["age"] = 20

print(dic)
```
