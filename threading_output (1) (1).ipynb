```
import time

def do_something():
    print("Sleeping 1 second...")
    time.sleep(1)
    print("Done sleeping")

start = time.perf_counter()  # record start time

do_something()

finish = time.perf_counter()
print(f"Finished in {round(finish-start,2)} second(s)")
```

```
import threading
import time

def do_something():
    print("Sleeping 1 second...")
    time.sleep(1)

start = time.perf_counter()

threads = []

for _ in range(10):
    t = threading.Thread(target=do_something)
    t.start()
    threads.append(t)

for t in threads:
    t.join()

finish = time.perf_counter()
print(f"Finished in {round(finish-start,2)} second(s)")
```

```
import threading
import time

def do_something(sec):
    print(f"Sleeping {sec} second(s)...")
    time.sleep(sec)

start = time.perf_counter()

threads = []

for i in range(1, 6):
    t = threading.Thread(target=do_something, args=[i])
    t.start()
    threads.append(t)

for t in threads:
    t.join()

finish = time.perf_counter()
print(f"Finished in {round(finish-start,2)} second(s)")
```

```
from concurrent.futures import ThreadPoolExecutor
import time

def do_something(sec):
    print(f"Sleeping {sec} second(s)...")
    time.sleep(sec)

start = time.perf_counter()

with ThreadPoolExecutor() as executor:
    secs = [1, 2, 3, 4, 5]
    results = executor.map(do_something, secs)

finish = time.perf_counter()
print(f"Finished in {round(finish-start,2)} second(s)")
```

```
import threading
import time
import random

def print_details(name, age):
    time.sleep(random.randint(1, 3))
    print(f"Name: {name}, Age: {age}")

names = ["hari", "ram", "sita", "john"]
ages = [20, 22, 24, 26]

threads = []

for name, age in zip(names, ages):
    t = threading.Thread(target=print_details, args=(name, age))
    t.start()
    threads.append(t)

for t in threads:
    t.join()
```

```
from concurrent.futures import ThreadPoolExecutor
import time

def task(sec):
    print(f"Task sleeping {sec} sec")
    time.sleep(sec)
    return f"Completed after {sec} sec"

start = time.perf_counter()

secs = [5, 4, 3, 2, 1]

with ThreadPoolExecutor() as executor:
    results = executor.map(task, secs)

for result in results:
    print(result)

finish = time.perf_counter()
print(f"Finished in {round(finish-start,2)} second(s)")
```

```
import threading
import requests
from pathlib import Path

def download_file(url, folder):
    folder.mkdir(exist_ok=True)
    filename = folder / url.split("/")[-1]

    print(f"Downloading {filename} ...")
    r = requests.get(url)
    filename.write_bytes(r.content)
    print(f"Downloaded {filename}")

urls = [
    "https://raw.githubusercontent.com/python/cpython/main/README.rst",
    "https://raw.githubusercontent.com/python/cpython/main/Lib/this.py",
    "https://raw.githubusercontent.com/python/cpython/main/Lib/random.py"
]

folder = Path("downloads")

threads = []

for url in urls:
    t = threading.Thread(target=download_file, args=(url, folder))
    t.start()
    threads.append(t)

for t in threads:
    t.join()

print("All downloads complete.")
```
