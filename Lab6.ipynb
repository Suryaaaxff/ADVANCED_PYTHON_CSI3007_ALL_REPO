```
def show_pet_details(species, name):
    """Display information about a pet."""
    print(f"Pet Name: {name}")
    print(f"Species : {species}")

show_pet_details("rabbit", "Oreo")
show_pet_details("turtle", "Shelly")
```

```
def display_pet(species, name):
    """Show pet information using keyword arguments."""
    print(f"{name} is a {species}")

display_pet(name="Milo", species="parrot")
display_pet(species="hamster", name="Whisky")
```

```
def introduce_pet(name, species):
    """Explain that order of parameters matters."""
    print(f"{name} belongs to the {species} category")

introduce_pet("Buddy", "dog")
introduce_pet("dog", "Buddy")
```

```
def pet_info(name, species="dog"):
    """Default species is dog unless changed."""
    print(f"{name} is a {species}")

pet_info("Rocky")
pet_info("Kiwi", "bird")
```

```
def register_pet(name, species="cat"):
    """Register pet with a default species."""
    print(f"Registered: {name} ({species})")

register_pet("Snowbell")
register_pet("Bolt", "dog")
```

```
def format_pet_name(name):
    """Capitalize and return cleaned pet names."""
    return name.strip().title()

print(format_pet_name(" brownie "))
print(format_pet_name("tommy"))
```

```
def greet_animals(animal_list):
    """Greet each animal in the list."""
    for animal in animal_list:
        print(f"Hello, {animal.title()}!")

animals = ["leo", "max", "bella"]
greet_animals(animals)
```

```
def create_pet_record(name, species):
    """Return a dictionary storing pet details."""
    return {"name": name, "species": species}

record = create_pet_record("Simba", "lion")
print(record)
```
